```python
#### Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.

## Here is a Python script to conduct a hypothesis test on the difference between two population means:

from scipy.stats import ttest_ind

def hypothesis_test(sample1, sample2):
    t_stat, p_val = ttest_ind(sample1, sample2)
    return t_stat, p_val


```

```python
#### Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom parameter set to 10.

# Here is a Python script to generate a Student's t-distribution plot:

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t

x = np.linspace(-5, 5, 100)
y = t.pdf(x, df=10)
plt.plot(x, y)
plt.title("Student's t-Distribution (df=10)")
plt.show()


```

```python
#### Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.

#Here is a Python program to calculate the two-sample t-test for independent samples:

from scipy.stats import ttest_ind

def two_sample_t_test(sample1, sample2):
    t_stat, p_val = ttest_ind(sample1, sample2)
    return t_stat, p_val

```

```python
#### Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50. Estimate the population mean revenue with a 95% confidence interval.

#To estimate the population mean revenue with a 95% confidence interval:

import math

sample_mean = 500
sample_std = 50
sample_size = 50

population_mean = sample_mean
margin_of_error = (sample_std / math.sqrt(sample_size)) * 1.96
confidence_interval = (population_mean - margin_of_error, population_mean + margin_of_error)
print(confidence_interval)


```

```python
#### Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

# Step 1: Formulate the null and alternative hypotheses

#Null Hypothesis (H0): μ = 10 mmHg (The new drug does not decrease blood pressure by 10 mmHg)
#Alternative Hypothesis (H1): μ < 10 mmHg (The new drug decreases blood pressure by less than 10 mmHg)

#Step 2: Calculate the test statistic

#t = (x̄ - μ) / (s / √n)
#= (8 - 10) / (3 / √100)
#= -2 / 0.3
#= -6.67

#Step 3: Determine the degrees of freedom

#df = n - 1 = 100 - 1 = 99

#Step 4: Find the critical value or p-value

#Using a t-distribution table or calculator, we find the critical value for a one-tailed test with df = 99 and α = 0.05:
#t_critical = -1.984

#Since our calculated t-statistic (-6.67) is less than the negative critical value (-1.984), we reject the null hypothesis.

#Alternatively, we can calculate the p-value using a t-distribution calculator or software:
#p-value ≈ 0.000001

#Since the p-value is less than α (0.05), we reject the null hypothesis.


```

```python
## Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5 pounds with a significance level of 0.01.

#Null Hypothesis (H0): μ = 5 pounds (The true mean weight is equal to 5 pounds)
#Alternative Hypothesis (H1): μ < 5 pounds (The true mean weight is less than 5 pounds)

#Step 2: Calculate the test statistic

#t = (x̄ - μ) / (s / √n)
#= (4.8 - 5) / (0.5 / √25)
#= -0.2 / 0.1
#= -2

#Step 3: Determine the degrees of freedom

#df = n - 1 = 25 - 1 = 24

#Step 4: Find the critical value or p-value

#Using a t-distribution table or calculator, we find the critical value for a one-tailed test with df = 24 and α = 0.01:
#t_critical = -2.492

#Since our calculated t-statistic (-2) is greater than the negative critical value (-2.492), we fail to reject the null hypothesis.

#Alternatively, we can calculate the p-value using a t-distribution calculator or software:
#p-value ≈ 0.021

#Since the p-value is greater than α (0.01), we fail to reject the null hypothesis.




```

```python
#### Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV program. They take a random sample of 50 viewers and find that the sample mean is 4 with a standard deviation of 1.5. Estimate the population mean with a 99% confidence interval.

##x̄ - (Z * (σ / √n)) < μ < x̄ + (Z * (σ / √n))

#where:
#x̄ = sample mean = 4
#σ = sample standard deviation = 1.5
#n = sample size = 50
#Z = Z-score corresponding to the desired confidence level (99% in this case) = 2.576

#First, calculate the margin of error (ME):

#ME = Z * (σ / √n)
#= 2.576 * (1.5 / √50)
#= 2.576 * 0.212
#= 0.547

#Now, construct the confidence interval:

#x̄ - ME < μ < x̄ + ME
#4 - 0.547 < μ < 4 + 0.547
#3.453 < μ < 4.547

#Therefore, the marketing company can be 99% confident that the average number of ads watched by viewers during the TV program is between 3.453 and 4.547.


```
